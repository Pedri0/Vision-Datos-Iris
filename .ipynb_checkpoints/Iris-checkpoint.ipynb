```python
import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix
import seaborn as sns

#Abrimos el archivo iris.data con la indicación de que no tiene encabezado
iris = pd.read_csv('iris.data',header = None)

# Hacemos un dataFrame con los datos recien abiertos.
Iris = pd.DataFrame(iris)

#añadimos nombres a los encabezados
#Los nombres de la columna se asignaron de acuerdo al archivo iris.names
Iris.columns = ['Longitud Sepalo','Ancho Sepalo','Longitud Petalo','Ancho Petalo','Especie']

#Imprimimos las primeras 10 filas
Iris.head(10)

```

```python
#Imprimimos las primeras diez filas y las primeras dos columnas
Iris[['Longitud Sepalo', 'Ancho Sepalo']].head(10)
```

```python
#Para contar el numero de filas y columnas usamos Iris.shape()
numero_filas = Iris.shape[0]
numero_columnas = Iris.shape[1]
print('El numero de filas es %d' %numero_filas + '. El numero de columnas es %d' %numero_columnas + '.')

print('\n')

#Para imprimir las llaves usamos la funcion keys()
print('Una llave en este DataFrame es ' + Iris.keys())

print('\n')

#Para encontrar Not a Number (NaN) usamos la funcion isnull()
print('El total de NaNs por columna son:')
print((Iris.isnull().sum()))
print('\n')
print('El total global de NaNs es: %d' %Iris.isna().values.sum() + '.')

print('\n')

#Creamos el array Solicitado con Numpy
arreglo = np.zeros((5,5),int)
print(arreglo)
print('\n')
#Rellenamos su diagonal con unos con la funcion fill_diagonal(x,y)
np.fill_diagonal(arreglo,1)
print(arreglo)
print('\n')
#Creamos la matriz dispersa con la funcion csr_matrix() de Scipy
dispersa = csr_matrix(arreglo)
print(dispersa)
print('El tipo de la matriz dispersa es: ' + str(type(dispersa)))
```

```python
Iris.describe()
```

El metodo describe() Muestra los valores de interes en los datos:
- count: Conteo de valores excluyendo los NaN
- mean: Media o promedio de todos los valores numericos en una columna
- std: Desviacion estandard de todos los valores numericos en una columna
- min: El valor numerico minimo de todos los valores numericos presentes en la columna
- 25%, 50%, 75% : Percentiles (cuantil 1, 2 y 3 respectivamente) de todos los valores numericos en una columna
- max: El valor numerico maximo de todos los valores numericos presentes en la columna

Tomado de https://backtobazics.com/python/pandas-describe-method-dataframe-summary/

```python
#Imprimimos solo el promedio
print('El promedio de las columnas por separado es:')
print(Iris.mean(axis=0,skipna=True))
print('\n')
print('Por Separado:')
print('El promedio de la columna Longitud Sepalo es: %f.' %Iris['Longitud Sepalo'].mean())
print('El promedio de la columna Ancho Sepalo es: %f.' %Iris['Ancho Sepalo'].mean())
print('El promedio de la columna Longitud Petalo es: %f.' %Iris['Longitud Petalo'].mean())
print('El promedio de la columna Ancho Petalo es: %f.' %Iris['Ancho Petalo'].mean())
print('\n')

#Imprimimos solo la desviacion
print('La desviacion de las columnas por separado es:')
print(Iris.std(axis=0,skipna=True))
print('\n')
print('Por Separado:')
print('La desviacion de la columna Longitud Sepalo es: %f.' %Iris['Longitud Sepalo'].std())
print('La desviacion de la columna Ancho Sepalo es: %f.' %Iris['Ancho Sepalo'].std())
print('La desviacion de la columna Longitud Petalo es: %f.' %Iris['Longitud Petalo'].std())
print('La desviacion de la columna Ancho Petalo es: %f.' %Iris['Ancho Petalo'].std())
```

```python
#Agrupamos por especie y contamos con las funciones groupby() y value_counts()
print(Iris.groupby('Especie')['Especie'].value_counts())

#Otro metodo mas sencillo es como sigue
Iris['Especie'].value_counts()
```
